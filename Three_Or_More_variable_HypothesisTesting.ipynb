Three or more Scale data -paired

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris

X,y = load_iris(return_X_y = True, as_frame = True)

#H0: There is no difference in sepal and petal measurements across flower types.
#H1: There is a difference in sepal and petal measurements across flower types.

X.head()
```

```python
y.unique()
```

```python
data = pd.concat([X,y],axis=1)
```

```python
data['target'] = data['target'].replace({0:'type_1', 1:'type_2', 2:'type_3'})
```

```python
data.columns
```

```python
#MANOVA - Multi variable Analysis
from statsmodels.multivariate.manova import MANOVA

data =data.rename(columns={
    'sepal length (cm)':'sepal_length', 'sepal width (cm)':'sepal_width', 'petal length (cm)':'petal_length',
       'petal width (cm)':'petal_width', 'target':'flower_name'

})
```

```python
data.columns
```

```python
result = MANOVA.from_formula('sepal_length + sepal_width + petal_length + petal_width ~ flower_name',data = data)
```

```python
print(result.mv_test())
```

```python
wilks_lamda = 0.0234 #Effect size
pvalue = 0.0000

#pvalue<0.05, We can reject the null hypothesis
#ur result is significant
#We can say that there is difference in the marks of students in each subject

if wilks_lamda > 0.9:
  print('small')
elif wilks_lamda >0.7:
  print('medium')
else:
  print('large')
```

```python
# The MANOAV test found that MANOVA.from_formula('sepal_length + sepal_width + petal_length + petal_width are not same for different flower name(pvalue <0.05)
# Futher wilkslamda effect size suggest the difference is large
```

Three or more ordinal data - paired

```python
stud_data = pd.read_csv('StudentStatistics.csv',sep = ';')
#H0: There is no difference in responses across teaching variables.
#H1: There is a difference in responses across teaching variables.
```

```python
stud_data.head()
```

```python
mydata = stud_data[['Teach_Motivate','Teach_LinkTheory','Teach_StimAsk','Teach_Avail']]
mydata
```

```python
mydata['Teach_Avail'].unique()
```

```python
#coding
coding={
    'Fully Disagree':0,
    'Disagree':1,
    'Neither disagree nor agree':2,
    'Agree':3,
    'Fully agree':4,
}

mydata.replace(coding, inplace=True)
mydata.dropna()
```

```python
#Using Friedman chi square test to find the pvalue
```

```python
mydata.dropna(inplace=True)
```

```python
import scipy
from scipy.stats import friedmanchisquare

friedmanchisquare(mydata['Teach_Avail'],mydata['Teach_LinkTheory'],mydata['Teach_StimAsk'],mydata['Teach_Motivate'])
```

```python
statistic=33.85869565217378
pvalue=2.1220432262951938e-07
```

```python
#Using Kendall W to find the effect size
n = mydata.shape[0]
k = mydata.shape[1]

w = statistic/(n*(k-1))
w
```

```python
if w <0.3:
  print('weak')
elif w < 0.5:
  print('Moderate')
elif w<0.7:
  print('relatively strong')
else:
  print('Strong')
```

```python
'''
The four ordinal variable were not having the same responses and after applying the friedman chi square
the difference was significant (pvalue < 0.05)
and futher the kendall w effect size suggest the difference is weak (0.212)
'''
```
